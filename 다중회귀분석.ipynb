## 이상치 검출

```python
def print_outlier(d):
    Q1 = d.quantile(0.25)

    Q3 = d.quantile(0.75)

    MAX = d.max()

    IQR = Q3 - Q1

    Lower_Fence = Q1 - (1.5 * IQR)

    Upper_Fence = Q3 + (1.5 * IQR)
    print('Lower_Fence : ',Lower_Fence)
    print('Upper_Fence : ',Upper_Fence)
    print('MAX : ',MAX)
    print('outlier : ',(Upper_Fence-Lower_Fence)/MAX)
    
def get_outlier(d):
    Q1 = d.quantile(0.25)

    Q3 = d.quantile(0.75)

    MAX = d.max()

    IQR = Q3 - Q1

    Lower_Fence = Q1 - (1.5 * IQR)

    Upper_Fence = Q3 + (1.5 * IQR)
    return (Upper_Fence-Lower_Fence)/MAX
```

```python
NN = 4
MM = 4
f, axes = plt.subplots(NN, MM)
f.set_size_inches((20, 30))
plt.subplots_adjust(wspace = 0.5, hspace = 0.3)
cols = data.columns[:-1]
i = 0
tmp = {}
for n in range(NN):
    for m in range(MM):
        axes[n,m].set_title(cols[i])
        sns.boxplot(y = cols[i], orient = "n", data = data, ax=axes[n,m])
        o = get_outlier(data[cols[i]])
        if o < 0.300:
            tmp[cols[i]] = [round(o,10)]
        i += 1
```
